```python
#Multipe Items
# Product 1
product1_name = input("Enter product_1 name: ")
count1 = int(input("Enter product_1 quantity: "))
price1 = float(input("Enter product_1 price: "))
gst1 = float(input("Enter product_1 gst %: "))

amount1 = count1 * price1
gst_amount1 = amount1 * gst1 / 100
total1 = amount1 + gst_amount1

# Product 2
product2_name = input("\nEnter product_2 name: ")
count2 = int(input("Enter product_2 quantity: "))
price2 = float(input("Enter product_2 price: "))
gst2 = float(input("Enter product_2 gst %: "))

amount2 = count2 * price2
gst_amount2 = amount2 * gst2 / 100
total2 = amount2 + gst_amount2

# Grand Total
grand_total = total1 + total2

print("\n===== BILL SUMMARY =====")
print(f"{product1_name}: Rs. {total1}")
print(f"{product2_name}: Rs. {total2}")
print(f"Grand Total: Rs. {grand_total:}")
```

```python
#Discount
product_name = input("Enter product name: ")
product_count = int(input("Enter product count: "))
price_per_item = float(input("Enter price per item: "))
discount_percentage = float(input("Enter discount percentage: "))
gst_percentage = float(input("Enter gst percentage: "))

amount_before_discount = product_count * price_per_item

discount_amount = amount_before_discount * discount_percentage / 100

amount_after_discount = amount_before_discount - discount_amount

gst_amount = amount_after_discount * gst_percentage / 100

final_amount = amount_after_discount + gst_amount

print("\n===== PRODUCT SUMMARY REPORT =====")
print(f"Product Name            : {product_name}")
print(f"Product Count           : {product_count}")
print(f"Price Per Item          : Rs. {price_per_item:}")
print(f"Amount Before Discount  : Rs. {amount_before_discount}")
print(f"Discount Percentage     : {discount_percentage}%")
print(f"Discount Amount         : Rs. {discount_amount}")
print(f"Amount After Discount   : Rs. {amount_after_discount}")
print(f"GST Percentage          : {gst_percentage}%")
print(f"GST Amount              : Rs. {gst_amount:}")
print(f"Final Amount            : Rs. {final_amount:}")
```
